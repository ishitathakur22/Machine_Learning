```python
#import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

#The warnings module in Python is used to alert you about things that aren't necessarily "errors"
#(which would stop your code), but rather "suggestions" or "notices" regarding the state of your code.
```

```python
df=pd.read_excel('customer_dataset.xlsx')
```

```python
df.head()
```

```python
#histogram with KDE =kernel density estimation 
#shows us an histogram with kde

plt.figure(figsize=(6,4))
sns.histplot(df['Spending_Score'],bins=10,kde=True,color='purple')#histogram wit kde overlay 
#spending score in the dataset
plt.title('distribution of spending score')
plt.xlabel('spending score')
plt.ylabel('frequency')
plt.show()
```

# scatter plot

```python
#sctter plot

plt.figure(figsize=(6, 4))
sns.scatterplot(x='Income', y='Spending_Score', data=df, hue='Gender', palette='deep', alpha=0.7)
plt.title('Income vs Spending Score') #scatter plot shiwing income vs spending score
plt.xlabel('Income')
plt.ylabel('Spending Score')
plt.show()
```

#  Box Plot

```python
# 3. Box Plot

plt.figure(figsize=(6, 4))
sns.boxplot(x='Region', y='Income', data=df, palette='pastel') # Box plot showing Income distribution
plt.title('Income Distribution by Region') # Box plot of Income by Region
plt.xlabel('Region')
plt.ylabel('Income')
plt.show()
```

#  Bar Plot

```python
# 4. Bar Plot

plt.figure(figsize=(6, 4))
sns.barplot(x='Education_Level', y='Income', data=df, estimator='mean', palette='muted') # Bar plot of Average Income
plt.title('Average Income by Education Level') # Bar plot of Average Income by Education Level
plt.xlabel('Education Level')
plt.ylabel('Average Income')
plt.xticks(rotation=45) # Rotate x-axis labels for better readability
plt.show()
```

#  Pair Plot

```python
# 5. Pair Plot

plt.figure(figsize=(5, 5))
sns.pairplot(df[['Age', 'Income', 'Spending_Score', 'Height', 'Weight', 'Gender']], diag_kind='kde', hue='Gender')
plt.show()
```

#  Heatmap 


```python
# 6. Heatmap (Correlation Matrix)

plt.figure(figsize=(6, 4))
corr_matrix = df[['Age', 'Income', 'Spending_Score', 'Height', 'Weight']].corr() # Compute the correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f') # Heatmap showing correlation between features
plt.title('Correlation Heatmap') # Heatmap of Correlation Matrix with annotations
plt.show()
```
