### Stemming

Stemming is the process of reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words known as a lemma. Stemming is important in natural language understanding (NLU) and natural language processing (NLP).

```python
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
words = ["eating", "eats", "eaten", "writing", "writes",
         "programming", "programs", "running", "studies", "playing"]
```

```python
for word in words:
    print(word+"--->"+stemming.stem(word))
```

```python
stemming.stem('congratulation')
```

```python
stemming.stem('sitting')
```

```python
lancaster = LancasterStemmer()
```

### RegexpStemmer

RegexpStemmer (Regular Expression Stemmer) is a stemmer in NLTK that removes parts of words based on a pattern (regular expression) that you define.

Unlike PorterStemmer or LancasterStemmer, it does not use built-in stemming rules. You tell it what to remove.

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer('ing$')
```

```python
print(reg_stemmer.stem('playing'))
print(reg_stemmer.stem('running'))
```

```python
from nltk.stem import RegexpStemmer

reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
print(reg_stemmer.stem('eating'))
```

```python
print(reg_stemmer.stem('running'))
```

### Snowball Stemmer

SnowballStemmer is an improved version of PorterStemmer. It is more accurate and supports multiple languages.

```python
from nltk.stem import SnowballStemmer
```

```python
snowballstemmer=SnowballStemmer('english',ignore_stopwords=False)
```

```python
for word in words:
    print(word+"--->"+snowballstemmer.stem(word))
```

### Lemmatization

Lemmatization is the process of converting a word to its actual dictionary form (lemma).

Unlike stemming, lemmatization produces meaningful words.

#### Wordnet Lemmatizer

Lemmatization technique is like stemming. The output we will get after lemmatization is called 'lemma', which is a root word rather than root stem, the output of stemming. After lemmatization, we will be getting a valid word that means the same thing.

NLTK provides WordNetLemmatizer class which is a thin wrapper around the WordNet corpus. This class uses morphy() function to the WordNet CorpusReader class to find a lemma. Let us understand it with an example.

Stemming → gives a root stem (may not be a real word)
studies → studi
caring → care
Lemmatization → gives a lemma (dictionary word)
studies → study
caring → care
better → good (with correct POS)

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lematizer= WordNetLemmatizer()
```

```python
for word in words:
    print(word + " ---> " + lemmatizer.lemmatize(word))
```

```python
lemmatizer.lemmatize("good",pos='v')
```


 "good" is an adjective, not a verb.
You specified:
pos='v'

which means treat the word as a verb.

Since WordNet cannot find a verb form of "good", it returns the word unchanged.

```python

lemmatizer.lemmatize("running", pos='v')
```
